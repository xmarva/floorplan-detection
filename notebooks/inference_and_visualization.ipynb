```python
import mmcv
from mmdet.apis import init_detector, inference_detector
import os
import matplotlib.pyplot as plt

config_file = '/kaggle/working/CubiCasa5k/mmdetection/configs/retinanet/custom_retinanet.py'
checkpoint_file = '/kaggle/working/CubiCasa5k/mmdetection/work_dirs/custom_retinanet/epoch_3.pth'

model = init_detector(config_file, checkpoint_file, device='cuda:0')
```

```python
import mmcv
import cv2
import matplotlib.pyplot as plt
import numpy as np
from mmdet.apis import init_detector, inference_detector
from mmdet.registry import VISUALIZERS

visualizer = VISUALIZERS.build(model.cfg.visualizer)
visualizer.dataset_meta = model.dataset_meta

img_path = '/kaggle/input/data-test/Screenshot 2024-09-20 at 2.25.42PM.png' 
img = cv2.imread(img_path)

result = inference_detector(model, img)

bboxes = result.pred_instances.bboxes.cpu().numpy()
labels = result.pred_instances.labels.cpu().numpy()
scores = result.pred_instances.scores.cpu().numpy()

for bbox, label, score in zip(bboxes, labels, scores):
    print(f"Bounding Box: {bbox}")
    print(f"Label: {label}")
    print(f"Score: {score}")
    print("---")

img_rgb = mmcv.imconvert(img, 'bgr', 'rgb')

visualizer.add_datasample(
    name='result',
    image=img_rgb,
    data_sample=result,
    draw_gt=False,  # Do not draw ground truth here
    pred_score_thr=0.3,  # Set threshold for showing predictions
    show=False  # Do not show immediately
)

img_result = visualizer.get_image()

img_result_rgb = mmcv.imconvert(img_result, 'bgr', 'rgb')

plt.figure(figsize=(10, 10))
plt.imshow(img_result_rgb)
plt.axis('off') 
plt.show()
```
